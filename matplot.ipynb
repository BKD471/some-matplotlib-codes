```python
x=np.linspace(-10,10,1000)
```

```python
y=1/(1+np.exp(-x))
```

```python
plt.plot(x,y,'r--')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'y--')
plt.subplot(2,2,4)
plt.plot(x,y,'b--')
```

```python
x=np.arange(0,10*np.pi,0.01)
y=np.sin(x)
z=np.cos(x)
w=np.tan(x)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,z,'g--')
plt.subplot(2,2,3)
plt.plot(x,w,'y--')
```
